## Match Result Table data

Importing the numpy, pandas for data manipulation, data wrangling and analysis. Requests library is used to make HTTP requests to the desired web pages and retrieve the HTML content. BeautifulSoup is best used for web scraping tasks that involve parsing and extracting information from static HTML pages and XML documents.

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# it requests a webpage from the specified URL and It retrieves the content of the webpage in text format
webpage = requests.get('https://www.espncricinfo.com/records/tournament/team-match-results/icc-cricket-world-cup-2023-24-15338').text
```

'lxml' refers to the lxml parser, which is a fast and feature-rich parser library for processing XML and HTML documents.
soup is object. 'soup' object allows you to navigate the HTML document in a more structured and accessible manner. you can use various BeautifulSoup methods and functions to extract specific elements, text, or attributes from the HTML.

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
headers = [] # empty list

# find all method to locate all span class with specified class attribute
head = soup.find_all('span', class_="ds-cursor-pointer")[1:8]

# loop to extract text without whitespaces from each span tag in list 'head'
for i in head:
  headers.append(i.text.strip())

headers

# creating a DataFrame with columns stored in list 'headers'
title = pd.DataFrame(columns = headers)
title

# empty list named as 'first_row' to store each span element located by soup object using method find all
first_row = []
row = soup.find_all('span', class_="")[8:]
for i in row:
  first_row.append(i.text.strip())

# first row contains complete table data in list format
# converting the list into array, to split the array into 48 parts(matches in the worldcup) using method 'array_split'
# the array list will have 48 list and each array will have data of individual match
def split_list(list1, parts):
    arr = np.array(list1)
    return np.array_split(arr, parts)

# Split the list into 4 equal parts
result = split_list(first_row, 48)

# converting each array into list using loop
for i in result:
  individual_list = i.tolist()

  # adding each converting list into a dataframe name - 'title'
  length = len(title)
  title.loc[length] = individual_list

title
```
